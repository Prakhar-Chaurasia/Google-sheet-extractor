```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
```

```python
scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_name('Google Sheet Extraction-5d83cb1ae1cf.json', scope)
client = gspread.authorize(creds)
```

```python
sheet = client.open('PENDING COLLECTIONS - FIELD TRACKER').sheet1
```

```python
data = sheet.get_all_records()
print(data)
```
